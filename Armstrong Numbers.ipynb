## Armstrong Numbers

### Problem :

Task:

Find out if a given number is an "**Armstrong Number**".

* An **n**-digit number that is the sum of the **n**th powers of its digits is called an **n-Armstrong** number. Examples :
<ol>
<li>371 = 3<sup>3</sup> + 7<sup>3</sup> + 1<sup>3</sup>;</li>
<li>9474 = 9<sup>4</sup> + 4<sup>4</sup> + 7<sup>4</sup> + 4<sup>4</sup>;</li>
<li>93084 = 9<sup>5</sup> + 3<sup>5</sup> + 0<sup>5</sup> + 8<sup>5</sup> + 4<sup>5</sup>.</li>
</ol>


* Write a Python program that;
<br>
<ol>
<li>takes a positive integer number from the user,</li>
<li>checks the entered number if it is Armstrong,</li>
<li>consider the negative, float and any entries other than numeric values then display a warning message to the user.</li>
</ol>

### Code :

```python
number = input("Enter a positive integer number : ")
sum = 0
try :
    if int(number) :
        if int(number) >= 0:
            # Assign the number received from the user to a temporary variable.
            temp_number = int(number)
            while temp_number > 0 :
                digit = temp_number % 10
                sum += digit ** len(number)
                temp_number //= 10
            if int(number) == sum :
                print(f"{number} is an Armstrong number")
            else :
                print(f"{number} is not an Armstrong number")
        else :
            print("It is an invalid entry. Don't use non-numeric, float, or negative values!")
except ValueError :
    print("It is an invalid entry. Don't use non-numeric, float, or negative values!")
```
